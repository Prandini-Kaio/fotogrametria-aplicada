```python
%cd /home/kaiooliveira/Documentos/kaio/dev/Fotogrametria/fotogrametria_aplicada/
import cv2
import matplotlib.pyplot as plt
from src import preprocessamento
import numpy as np

K = np.array([
    [2559.68, 0, 1536],
    [0, 2559.68, 1152],
    [0, 0, 1]
])

dist = np.array([-0.0204997, 0, 0, 0, 0]) # apenas k1, os demais sao 0

img_path = "/home/kaiooliveira/Documentos/kaio/dev/Fotogrametria/fotogrametria_aplicada/data/bruto/south-building/images/P1180161.JPG" 
img_proc = preprocessamento.processar_imagem(img_path, K, dist)

# Mostra original vs processada
img_original = cv2.imread(img_path)
img_original = cv2.cvtColor(img_original, cv2.COLOR_BGR2RGB)
img_proc = cv2.cvtColor(img_proc, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(12, 15))
plt.subplot(1, 2, 1)
plt.title("Imagem Original")
plt.imshow(img_original)
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Imagem Processada")
plt.imshow(img_proc)
plt.axis("off")

plt.show()
```
